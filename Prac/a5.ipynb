```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("ML Dataset/Regression/diabetes.csv")

# Handle missing values (if any)
data = data.fillna(data.mode().iloc[0])

# Select one feature for simple regression
X = data[['DiabetesPedigreeFunction']]
y = data['Outcome']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Residual Sum of Squares (RSS)
rss = np.sum((y - y_pred) ** 2)

# Coefficient of Determination (R²)
r2 = r2_score(y, y_pred)

# Print results
print("Coefficient (slope):", slope)
print("Intercept:", intercept)
print("Residual Sum of Squares (RSS):", rss)
print("Coefficient of Determination (R²):", r2)

# Plot the regression line
plt.figure(figsize=(8,6))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('Diabetes Pedigree Function')
plt.ylabel('Outcome')
plt.title('Simple Linear Regression: Diabetes Dataset')
plt.legend()
plt.show()
```
